```
####Question 24####
####Какая студия даёт самые длинные названия своим фильмам по количеству символов?
movie = pd.read_csv('movie_bd_v5.csv') #импортируем данные imdb
movie['production'] = movie.production_companies.str.split(pat = '|') #разделяем каждую ячейку переменной production по знаку '|' и трансформуруем ее в list
movie = movie.explode('production') #разбиваем каждую ячейку переменной production и растягаваем датасет movie

movie['name_len'] = movie.original_title.apply(len) #сохраняем длинну названия фильма в переменную name_len
movie.groupby(['production'])['name_len'].mean().sort_values(ascending=False) #группируем фильмы по студиям, считаем среднее значение переменной name_len и выводим в порядке убывания 

#Answer: Four By Two Productions 
```

```
####Question 27####
####Какие актеры чаще всего снимаются в одном фильме вместе?
import itertools as iter #импортируем библиотеку itertools
import collections #импортируем библиотеку 

movie = pd.read_csv('movie_bd_v5.csv') #импортируем данные imdb
movie['cast'] = movie.cast.str.split(pat = '|') #разделяем каждую ячейку переменной cast по знаку '|' и трансформуруем ее в list

movie['pairs'] = movie['cast'].apply(lambda x: list(iter.combinations(x, 2))) # создаем новую переменную pair, в которую сохраняем все возможные сомбинации из двух элеменов ячейки cast
movie = movie.explode('pairs') #разбиваем каждую ячейку переменной pair и растягаваем датасет movie

collections.Counter(movie['pairs']).most_common(1) #находим самый часто встречаемый элемент с помощью библиотекии collections

#Answer: [(('Daniel Radcliffe', 'Rupert Grint'), 8)]
```
